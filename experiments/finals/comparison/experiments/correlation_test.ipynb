```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
```

```python
path_dataset = "/home/diego-ngz/Git/thesis-tabtrans/Final_models_4/all_datasets/balanced_accuracies.csv"
# Load the CSV into a pandas DataFrame
data = pd.read_csv(path_dataset)
```

```python
# Assuming `data_melted` is your DataFrame
tabtrans_df = data[data['model'] == 'balanced_accuracy_tabtrans']
```

```python
tabtrans_df =tabtrans_df.rename(columns={"balance_accuracy": "balanced_accuracy"})
```

```python
tabtrans_df.head()
```

```python
#Check linearity of the data vs balanced accuracy
# Scatter 
x = tabtrans_df["n_instances"]
y = tabtrans_df["balanced_accuracy"]
plt.scatter(x,y)
plt.xlabel('# Instances')
plt.ylabel('Balanced Accuracy')
plt.title('Scatter Plot: # Instances vs. Balanced Accuracy')
plt.show()


# Perform Spearman's rank correlation
corr, p_value = spearmanr(x, y)

print(f"Spearman's correlation coefficient: {corr}")
print(f"P-value: {p_value}")
```
